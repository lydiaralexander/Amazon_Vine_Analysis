```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Books_v1_02.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
# Create the vine_table. DataFrame
vine_df = df.select(['review_id', 'star_rating', 'helpful_votes', 'total_votes', 'vine', 'verified_purchase'])
vine_df.show()
```

```python
#filter data and create new dataframe
total_votes_df = vine_df.filter("total_votes>=20")
total_votes_df.show()
total_votes_df.count()
```

```python
helpful_votes_df = total_votes_df.filter("(helpful_votes/total_votes) >= .5")
helpful_votes_df.show()
helpful_votes_df.count()
```

```python
paid_review_df = helpful_votes_df.filter("vine == 'Y'")
paid_review_df.show()
paid_review_df.count()
```

```python
non_paid_review_df = helpful_votes_df.filter("vine == 'N'")
non_paid_review_df.show()
non_paid_review_df.count()
```

```python
total_reviews = helpful_votes_df.count()
print("Total Reviews:", total_reviews)

five_star_reviews = helpful_votes_df.filter("star_rating == 5").count()
print("Total Five Star Reviews:", five_star_reviews)


five_star_and_paid = helpful_votes_df.filter("star_rating = 5 AND vine == 'Y'").count()
print("Paid Five Star Reviews:", five_star_and_paid)

five_star_and_unpaid = helpful_votes_df.filter("star_rating = 5 AND vine == 'N'").count()
print("Unpaid Five Star Reviews:", five_star_and_unpaid)

percentage_paid_five_star = ((five_star_and_paid/five_star_reviews)*100)
print("Percentage of Five Star and Paid Reviews:", percentage_paid_five_star,"%")

percentage_unpaid_five_star = ((five_star_and_unpaid/five_star_reviews)*100)
print("Percentage of Five Star and Unpaid Reviews:", percentage_unpaid_five_star,"%")
```
